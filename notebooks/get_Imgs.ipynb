```python
import cv2
import os
from tqdm import tqdm
```

```python
video_path = "../exp-data/20251201/28/10V_10X_5_2025-12-01T17-55-02.309.avi"
output_dir = "../exp-data/20251201/28/10_5"
os.makedirs(output_dir, exist_ok=True)

cap = cv2.VideoCapture(video_path)
total_frames = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
fps = 100

frame_count = 0
ini_frame, end_frame = 0, total_frames
crop_size = 1000 

with tqdm(total=total_frames, desc="Saving frames", unit="frames") as pbar:
    while frame_count < end_frame:
        ret, frame = cap.read()
        if not ret:
            break
        if frame_count >=ini_frame:
            h, w = frame.shape[:2]
            if h < crop_size or w < crop_size:    
                raise ValueError("Frame is smaller than 1000px-by-1000px")
    
            # cropping...
            cropped = frame[0:crop_size, 0:crop_size]
    
            # Saving...
            f_name = os.path.join(output_dir, f"{frame_count/fps:.2f}s.tif")
            cv2.imwrite(f_name, cropped)

        frame_count += 1
        pbar.update(1)

cap.release()
```
