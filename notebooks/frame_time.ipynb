```python
import numpy as np
import random as rd

import sys
sys.path.append('../scripts/')
from frame_time import corrected_time
```

```python
drop_rate = np.average([28.06, 29.08, 30.56])
```

```python
start_frame = 100
fps = 100
end_frame = 1000

ts = list(range(start_frame, int(end_frame * 100/(100-drop_rate)), 1))

n_samples = 1000
t_samples = np.array([sorted(rd.sample(ts, end_frame)) for i in range(n_samples)])

pos_mean = np.median(t_samples, axis=0)/fps
pos_std  = np.std(t_samples, axis=0)/fps
# pos_mean, pos_std
```

```python
import matplotlib.pyplot as plt
plt.plot( pos_mean, pos_std)
# plt.errorbar(pos_mean, list(range(0, len(pos_mean))), xerr=pos_std)
```
