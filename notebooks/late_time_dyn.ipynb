```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
import json
```

```python
matplotlib.rcParams['axes.titlesize'] = 20
matplotlib.rcParams['axes.labelsize'] = 20
```

```python
with open('../results/10V_10X_processed.json', 'r') as f1:
    t_10, S_dyn_10 = np.transpose(json.load(f1))

with open('../results/20V_10X_processed.json', 'r') as f2:
    t_20, S_dyn_20 = np.transpose(json.load(f2))

with open('../results/40V_10X_processed.json', 'r') as f3:
    t_40, S_dyn_40 = np.transpose(json.load(f3))

with open('../results/60V_10X_processed.json', 'r') as f4:
    t_60, S_dyn_60 = np.transpose(json.load(f4))
```

```python
fig = plt.figure(figsize=(12, 8))

plt.plot(np.log(t_10[:]), np.log(S_dyn_10[:]), label = '10V')
plt.plot(np.log(t_20[:]), np.log(S_dyn_20[:]), label = '20V')
plt.plot(np.log(t_40[:]), np.log(S_dyn_40[:]), label = '40V')
plt.plot(np.log(t_60[:]), np.log(S_dyn_60[:]), label = '60V')
plt.grid()
plt.xlabel('log(t (s))')
plt.ylabel(r'$log(\rho\ (mm^{-2}) \times Volume)$')
plt.legend()
plt.savefig('../figs/raw-combind-rho_dyn.png')
```

## Late Time Coarsening Dynamics

```python
print(np.where(np.log(S_dyn_10[:]) == np.max(np.log(S_dyn_10[:]))))
print(np.where(np.log(S_dyn_20[:]) == np.max(np.log(S_dyn_20[:]))))
print(np.where(np.log(S_dyn_40[:]) == np.max(np.log(S_dyn_40[:]))))
print(np.where(np.log(S_dyn_60[:]) == np.max(np.log(S_dyn_60[:]))))
```

```python
fig = plt.figure(figsize=(12, 8))

a, b, c = 117, 164, 368 
t_10_ = t_10 - (t_10[a] -1)
plt.plot(np.log(t_10_[a:b]), np.log(S_dyn_10[a:b]), label = '10V (Defect-formation)')
plt.plot(np.log(t_10_[b:c]), np.log(S_dyn_10[b:c]), label = '10V (Late-time)')

slope_, intersect_ = np.polyfit(np.log(t_10_[b:c]), np.log(S_dyn_10[b:c]), deg=1)
plt.title('slope(Late-time) = {:.2f}'.format(slope_))
plt.plot(np.log(t_10_[b:c]), slope_ * np.log(t_10_[b:c]) + intersect_, label = 'Late-time fit')


plt.grid()
plt.xlabel('log(t (s))')
plt.ylabel(r'$log(\rho\ (mm^{-2}) \times Volume)$')
plt.legend()
plt.savefig('../figs/10V-rho_dyn.png')
```
