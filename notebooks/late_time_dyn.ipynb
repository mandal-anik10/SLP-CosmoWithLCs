```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
import json
```

```python
matplotlib.rcParams['axes.titlesize'] = 20
matplotlib.rcParams['axes.labelsize'] = 20
```

```python
with open('../results/10V_10X_processed.json', 'r') as f1:
    t_10, S_dyn_10 = np.transpose(json.load(f1))

with open('../results/20V_10X_processed.json', 'r') as f2:
    t_20, S_dyn_20 = np.transpose(json.load(f2))

with open('../results/40V_10X_processed.json', 'r') as f3:
    t_40, S_dyn_40 = np.transpose(json.load(f3))

with open('../results/60V_10X_processed.json', 'r') as f4:
    t_60, S_dyn_60 = np.transpose(json.load(f4))
```
