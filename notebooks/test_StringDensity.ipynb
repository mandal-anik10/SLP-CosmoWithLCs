```python
import numpy as np
import matplotlib.pyplot as plt
from skimage import io, color, filters, morphology, measure

# 1. Load and convert to grayscale
img = io.imread('Pictures/Screenshots/Screenshot from 2025-09-18 15-44-07.png')
gray = color.rgb2gray(color.rgba2rgb(img))

# 2. Denoise & enhance contrast (optional)
gray = filters.gaussian(gray, sigma=1)

# 3. Threshold to binary (defect lines appear dark)
th = filters.threshold_otsu(gray)
bw = gray < th

# 4. Skeletonize to single-pixel width
skel = morphology.skeletonize(bw)

# 5. Compute total length in pixels
length_pixels = np.count_nonzero(skel)

# 6. Convert to µm (scale: s µm/pixel)
s = 0.1  # example: 0.1 µm per pixel
L_total = length_pixels * s

# 7. Compute imaged area in µm²
H, W = gray.shape
A = (W * s) * (H * s)

# 8. String density
rho_s = L_total / A  # µm per µm² = 1/µm

print(f"Total length: {L_total:.2f} µm")
print(f"Area: {A:.0f} µm²")
print(f"String density: {rho_s:.4f} µm⁻¹")
plt.imshow(img)
```
