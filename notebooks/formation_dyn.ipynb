```python
Sden2_avg = np.array(np.average([Sden2_1[:25], Sden2_2[:25], Sden2_3[:25]], axis=0))
Sden2_sd = np.array(np.std([Sden2_1[:25], Sden2_2[:25], Sden2_3[:25]], axis=0))
Sden3_avg = np.array(np.average([Sden3_1[:25], Sden3_2[:25], Sden3_3[:25]], axis=0))
Sden3_sd = np.array(np.std([Sden3_1[:25], Sden3_2[:25], Sden3_3[:25]], axis=0))
Sden4_avg = np.array(np.average([Sden4_1[:25], Sden4_2[:25], Sden4_3[:25]], axis=0))
Sden4_sd = np.array(np.std([Sden4_1[:25], Sden4_2[:25], Sden4_3[:25]], axis=0))
Sden5_avg = np.array(np.average([Sden5_1[:25], Sden5_2[:25], Sden5_3[:25]], axis=0))
Sden5_sd = np.array(np.std([Sden5_1[:25], Sden5_2[:25], Sden5_3[:25]], axis=0))
Sden6_avg = np.array(np.average([Sden6_1[:25], Sden6_2[:25], Sden6_3[:25]], axis=0))
Sden6_sd = np.array(np.std([Sden6_1[:25], Sden6_2[:25], Sden6_3[:25]], axis=0))
Sden7_avg = np.array(np.average([Sden7_1[:25], Sden7_2[:25], Sden7_3[:25]], axis=0))
Sden7_sd = np.array(np.std([Sden7_1[:25], Sden7_2[:25], Sden7_3[:25]], axis=0))
Sden8_avg = np.array(np.average([Sden8_1[:25], Sden8_2[:25], Sden8_3[:25]], axis=0))
Sden8_sd = np.array(np.std([Sden8_1[:25], Sden8_2[:25], Sden8_3[:25]], axis=0))
Sden9_avg = np.array(np.average([Sden9_1[:25], Sden9_2[:25], Sden9_3[:25]], axis=0))
Sden9_sd = np.array(np.std([Sden9_1[:25], Sden9_2[:25], Sden9_3[:25]], axis=0))
```

```python
plt.plot(t2, Sden2_avg, label='200Vpp')
plt.plot(t3, Sden3_avg, label='300Vpp')
plt.plot(t4, Sden4_avg, label='400Vpp')
plt.plot(t5, Sden5_avg, label='500Vpp')
plt.plot(t6, Sden6_avg, label='600Vpp')
plt.plot(t7, Sden7_avg, label='700Vpp')
plt.plot(t8, Sden8_avg, label='800Vpp')
plt.plot(t9, Sden9_avg, label='900Vpp')
plt.legend()
```

```python
effectiv_cell_area = 8e-5    # m^2
C = 16.23e-12    # f
cell_thickness = (effectiv_cell_area * 8.85e-12 / C)
cell_thickness
ramp_rate = ((np.linspace(0.2,  0.9, 8)/cell_thickness)/4e4) 
ramp_rate
```

```python
fig, axs = plt.subplots(2, 2, figsize=(25, 15))
ax = axs.flatten()
# plt.tight_layout()

bb = [2, 3, 4, 5]
for i in range(4):
    b = bb[i]
    t = np.median([t2[b], t3[b], t4[b], t5[b], t6[b], t7[b], t8[b], t9[b]])
    defects_formed = np.array([Sden2_avg[b], Sden3_avg[b], Sden4_avg[b], Sden5_avg[b], Sden6_avg[b], Sden7_avg[b], Sden8_avg[b], Sden9_avg[b]])
    defects_formed_err = np.array([Sden2_sd[b], Sden3_sd[b], Sden4_sd[b], Sden5_sd[b], Sden6_sd[b], Sden7_sd[b], Sden8_sd[b], Sden9_sd[b]])
    
    ax[i].errorbar(ramp_rate, defects_formed, yerr=defects_formed_err, fmt='o--', label=r'Mean density $\pm1\sigma$ error')
    ax[i].set_title('{:.2f}s'.format(t))
    ax[i].set_xlabel(r'Electric field ramp rate ($Vm^{-1}s^{-1}$)')
    ax[i].set_ylabel(r'Defect Density ($mm^{-2}$)')
    ax[i].set_xlim([0.55, 0.1])
    ax[i].set_xscale('log')
    ax[i].set_yscale('log')
    ax[i].legend()
    ax[i].grid()
plt.savefig('../figs/formation_dyn.png')
```
